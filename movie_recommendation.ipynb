'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
we want to convert above list to this format
['Action','Fantasy','Science Fiction']

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

given list is a string of list so we have to convert it to int of list to do this we use ast.literal_eval() function

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies
```

```python
movies['cast']
```

```python
#we only want three cast
def convertcast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
        
    return L
```

```python
movies['cast']=movies['cast'].apply(convertcast)
```

```python
movies
```

```python
movies.iloc[0].crew
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

```python
#convertin all to lower case
new_df['tag']=new_df['tag'].apply(lambda x:x.lower())
```

# now we do vectorization of tag to find similar movies
nearest vector will be moe similar in nature
text to vector using bag of word

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors.shape
```

# now calculating distance between vector using cosine distance we will not go with euclidean dist bcoz it gives some time error


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#sorted(similarity[0],reverse=True) we cannot use this bcoz of this sorting we will loose our indexing of movies
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] #finding index of entered movie
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(i[0])
    
    
```

```python
recommend('Avatar')
```
